```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import patches as mpt
```

## Used data 
I've already mentioned about the used data. Down below is how Data frame looks like:

```python
chocolate_data = pd.read_csv("/kaggle/input/chocolate-sales/Chocolate Sales (2).csv")
chocolate_data
```

# About the countries

```python
fig, ax1 = plt.subplots()
countries = chocolate_data["Country"].value_counts()
ax1.pie(countries, labels=countries.index, autopct='%1.0f%%')
ax1.set_title("Chocolate's producers distribution around the world")
plt.show()
```

Despite Austrialia being country with second lowest population in this list of countries, it has the biggest share in chocolate's production in the world.

```python
product = chocolate_data["Product"].value_counts()
print(product)
fix, (ax1, ax2) = plt.subplots(2,1)
colors = ["lime" if i >= 150 else "yellow" if i>=130 else "red" for i in product.values]
ax1.bar(product.index, product.values, width=1, color=colors)
ax1.set_xticklabels(product.index, rotation=45, ha="right", size = 15)
color_lime = mpt.Patch(color="lime", label="total >= 150")
color_yellow = mpt.Patch(color="yellow", label="150 > total >= 130")
color_red = mpt.Patch(color="red", label="total < 130")
ax1.legend(handles=[color_lime, color_yellow, color_red], fontsize=20)


ax2.pie(product, labels=product.index, autopct='%1.0f%%')
plt.subplots_adjust(left=0, right=3, top=7, bottom=0)
plt.show()
```
